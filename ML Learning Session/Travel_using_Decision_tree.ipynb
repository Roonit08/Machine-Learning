```python
travel.drop(columns=['CustomerID'],inplace=True)
travel.drop(columns=['ProductPitched'],inplace=True)
```

```python
## median is used to fill the data with na values present here i am ronit bhattarai and i did this reviewing it with my older data set titanic
travel['Age'].fillna(travel['Age'].median(), inplace=True)
travel['DurationOfPitch'].fillna(travel['DurationOfPitch'].median(), inplace=True)
travel['NumberOfPersonVisiting'].fillna(travel['NumberOfPersonVisiting'].median(), inplace=True)
travel['NumberOfFollowups'].fillna(travel['NumberOfFollowups'].median(), inplace=True)
travel['PreferredPropertyStar'].fillna(travel['PreferredPropertyStar'].median(), inplace=True)
travel['NumberOfTrips'].fillna(travel['NumberOfTrips'].median(), inplace=True)
travel['PitchSatisfactionScore'].fillna(travel['PitchSatisfactionScore'].median(), inplace=True)
travel['MonthlyIncome'].fillna(travel['MonthlyIncome'].median(), inplace=True)
travel['NumberOfChildrenVisiting'].fillna(travel['NumberOfChildrenVisiting'].median(), inplace=True)
```

```python
travel.isnull().sum()
```

```python
## i filled the missing value with median above and for type of contatct i used mode
travel['TypeofContact'].fillna(travel['TypeofContact'].mode()[0], inplace=True)
```

```python
travel.isnull().sum()
```

```python
type_contact = pd.get_dummies(travel['TypeofContact'], drop_first=True)
occupation = pd.get_dummies(travel['Occupation'], drop_first=True)
gender = pd.get_dummies(travel['Gender'], drop_first=True)
marital = pd.get_dummies(travel['MaritalStatus'], drop_first=True)
designation = pd.get_dummies(travel['Designation'], drop_first=True)
```

```python
travel.drop(['TypeofContact','Occupation','Gender','MaritalStatus','Designation'],axis=1,inplace=True)
```

```python
travel = pd.concat([travel, type_contact, occupation, gender, marital, designation],axis=1)
```

```python
travel.dtypes
```

```python
travel.isnull().sum()
```

```python
travel
```

```python
x=travel.drop('ProdTaken',axis=1)
y=travel['ProdTaken']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier=DecisionTreeClassifier()
```

```python
tree_classifier.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier,filled=True)
```

```python
prediction=tree_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm = confusion_matrix(y_test,prediction)
print(cm)
```

```python
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens")
```

```python
print(classification_report(y_test,prediction))
print("The accuracy of this model is 89%.")
```
