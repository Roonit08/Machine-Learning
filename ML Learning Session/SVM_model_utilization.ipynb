```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

```python
cancer['data']
```

```python
df_feature = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df_feature
```

```python
cancer['target_names']
```

```python
df_target=pd.DataFrame(cancer['target'],columns=['is_malignant_cancer'])
df_target
```

```python
df_feature.describe()
```

```python
df_feature.info()
```

```python
df_feature.isnull().sum()
```

```python
x=df_feature
```

```python
y=df_target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.svm import SVC
```

```python
model= SVC()
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
print("The accuracy of this model is 91%.")
```

```python
cm=confusion_matrix(y_test,prediction)
cm
```

```python
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
```

```python
from sklearn.metrics import accuracy_score
```

```python
linear_svc=SVC(kernel='linear',C=10.0)
linear_svc.fit(x_train,y_train)
prediction=linear_svc.predict(x_test)
prediction
```

```python
accurate=(accuracy_score(y_test,prediction)*100)
print("The accuracy is:",accurate,"%")
```

```python
linear_svc.score(x_train,y_train)
```

```python
sigmoid_svc=SVC(kernel='sigmoid',C=10.0)
sigmoid_svc.fit(x_train,y_train)
prediction=sigmoid_svc.predict(x_test)
prediction
```

```python
sigmoid_svc.score(x_train,y_train)
```

```python
accurate=(accuracy_score(y_test,prediction)*100)
print("The accuracy is:",accurate,"%")
```

```python
rbf_svc=SVC(kernel='rbf',C=12.0)
rbf_svc.fit(x_train,y_train)
prediction=rbf_svc.predict(x_test)
prediction
```

```python
accurate=(accuracy_score(y_test,prediction)*100)
print("The accuracy is:",accurate,"%")
```

```python
rbf_svc.score(x_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
para_grid={
    'C'  :[1,2,4,6,10],
    'kernel' :['rbf','linear']
}
opt_model=GridSearchCV(SVC(),para_grid)
opt_model.fit(x_train,y_train)
prediction_opt=opt_model.predict(x_test)
```

```python
opt_model.best_params_
```

```python
most_accurate=opt_model.score(x_train,y_train)
print("The most accurate model is linear with:",most_accurate*100,"% accuracy score.")
```

```python
query=df_feature.loc[100:100]
query
```

```python
result=opt_model.predict(query)
if result == 1:
    print("Malignant")
else:
    print("Belign")
```
