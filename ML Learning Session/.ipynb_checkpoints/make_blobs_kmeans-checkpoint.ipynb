```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y = make_blobs(n_samples=300,centers=3,n_features=2)
```

```python
x,y
```

```python
data= pd.DataFrame(x, columns=['Feature_1', 'Feature_2'])
data
```

```python
data.info()
```

```python
data.columns
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x='Feature_1',y='Feature_2',data=data)
```

```python
X= data[['Feature_1','Feature_2']]
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
km.fit(X)
```

```python
prediction = km.predict(X)
prediction
```

```python
data['cluster']=prediction
data['cluster']
```

```python
data1=data[data['cluster']==0]
data2=data[data['cluster']==1]
```

```python
plt.scatter(data1['Feature_1'],data1['Feature_2'],color="Green")
plt.scatter(data2['Feature_1'],data2['Feature_2'],color="Red")
```

```python
sse=[]
for k in range(1,11):
    km=KMeans(n_clusters=k)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
plt.plot(range(1,11),sse)
plt.xlabel("Number of clusters")
plt.ylabel("sse")
plt.show()
```
