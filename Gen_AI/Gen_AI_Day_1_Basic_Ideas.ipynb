```python
from groq import Groq
GROQ_API_KEY = "gsk_xxxxxxxxxxxxxxxxxxxxxxx" 

client = Groq (api_key = GROQ_API_KEY)
```

```python
system_prompt = {
    "role" : "system",
    "content" : "You' re very helpful and funny AI. Your name is jarvis named by Ronit Bhattarai. Whatever user asks, you have to answer in wise way."
}
user_query ={
    "role" : "user",
    "content": "What is a Computer?"
}
print(f"User : {user_query["content"]}")
response = client.chat.completions.create(
    model = "llama-3.1-8b-instant",
    messages = [system_prompt, user_query],
    max_tokens = 1500,
    temperature = 0.7
)
    
result = response.choices[0].message.content
print(f"AI : {result}")
```
