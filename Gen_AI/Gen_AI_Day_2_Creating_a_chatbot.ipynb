```python
from groq import Groq
GROQ_API_KEY = "gsk_xxxxxxxxxxxxxxxxxxxxx"  # how to use env here

client = Groq (api_key = GROQ_API_KEY)
```

```python
system_prompt = {
    "role": "system",
    "content": "You' re very helpful and funny AI. Your name is jarvis named by Ronit Bhattarai. Whatever user asks, you have to answer in wise way."
}

chat_history = [system_prompt]

while True:
    user_input = input("You: ")
    user_query = {
        "role": "user",
        "content": user_input
    }
    chat_history.append(user_query)
    
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=chat_history,
        max_tokens=1500,
        temperature=0.7
    )
    result = response.choices[0].message.content
    print(f"AI: {result}")

    
    ai_response_context = {
        "role": "assistant",
        "content": result
    }
    chat_history.append(ai_response_context)
```
