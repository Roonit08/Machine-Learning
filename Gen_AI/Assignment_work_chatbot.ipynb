```python
# importing API data
from groq import Groq
from dotenv import load_dotenv
import os

load_dotenv()
API_KEY = os.getenv("Assignment_work_chatbot_API_key")
client = Groq(api_key=API_KEY)
```

```python
# setting up system prompt for the chatbot
system_prompt ={
    "role": "system",
    "content": "you're very helpful and friendly assistant. your name is ROO. Whatever user asks you, you will answer in a very friendly and helpful manner. you will answer in a very simple and easy words."
}

chat_history = [system_prompt]

while True:
    user_input = input("You: ")
    user_query = {
        "role":"user",
        "content": user_input
    }
    chat_history.append(user_query)

    response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=chat_history,
    max_tokens=1500,
    temperature=0.7
    )
    result = response.choices[0].message.content
    print(f"ROO: {result}")

    ai_response_context = {
    "role": "assistant",
    "content": result
    }
    chat_history.append(ai_response_context)
```
