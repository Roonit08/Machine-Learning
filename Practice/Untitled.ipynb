```python
for index in df.index:
   if df.loc[index,'Survived']==0:
      df.loc[index,'Survived']='dead'
   else:
     df.loc[index,'Survived']='alive'  
```

```python
df
```

```python
def change_survived(a):
    if a == 0:
        return 'dead'
    else:
        return 'alive'
```

```python
change_survived(0) #calling a function
```

```python
df.Survived = df.Survived.apply(change_survived)
```
