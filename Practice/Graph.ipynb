```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
a= np.array([1,2,3,4,5,6,7])
```

```python
b=a**2
```

```python
a
```

```python
b
```

```python
plt.plot(a,b)
plt.show()
```

```python
plt.plot(a,b)
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,a)
plt.plot(a,b)
plt.plot(a,b)
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,b,marker='o')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,b,marker='d')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,b,marker='o', linestyle='dotted')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,b,marker='o',linestyle='dotted',color='r')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,b,marker='o',linestyle='dotted',color='r',ms='15')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.show()
```

```python
plt.plot(a,b,marker='o',linestyle='dotted',color='r')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.grid(axis='x')
plt.show()
```

```python
plt.plot(a,b,marker='o',linestyle='dotted',color='r')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.grid(axis='y')
plt.show()
```

```python
plt.plot(a,b,marker='o',linestyle='dotted',color='r')
plt.xlabel('Day')
plt.ylabel('Price')
plt.title('Day vs Price')
plt.grid(axis='y')
plt.grid(axis='x')
plt.show()
```
