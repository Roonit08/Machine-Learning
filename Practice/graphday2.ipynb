```python
S=['male','female']
male = (df.Sex=='male').sum()
female = (df.Sex=='female').sum()
```

```python
plt.bar(['male','female'],[male,female])
plt.title('Total population')
plt.xlabel('number')
plt.ylabel('gender')
plt.show()
```

```python
plt.pie([male,female],labels=S,autopct ='%.2f%%')
plt.show()
```

```python
dead= (df.Survived=='0').sum()
alive = (df.Survived=='1').sum()
```

```python
plt.bar(['dead','alive'],[dead,alive],)
plt.title('Survival Bar')
plt.xlabel('number')
plt.ylabel('survival')
plt.show()
```
